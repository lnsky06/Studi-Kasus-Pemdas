# Studi Kasus 1 :
Rumah makan padang memiliki menu yang enak. Mereka ingin memotong beberapa item dari menu. Untuk membuat sebagian besar pelanggan mereka senang, mereka ingin menghapus hanya 3 menu makanan yang paling tidak populer dari menu. Kita memiliki menu makanan yang disajikan dalam rumah makan tersebut, serta hitungan berapa kali makanan tersebut terjual dalam seminggu terakhir. Data dapat dilihat dibawah ini!

masakan_padang = {

'Rendang':95,

'Dendeng':93,

'Ayam Bakar':90,

'Ayam Gulai':92,

'Gulai Tunjang':90,

'Gulai Pakis':83,

'Kikil':85,

'Ikan Tongkol Balado':89,

'Telor Balado':87,

'Perkedel':84 }

Tugas :

Buat visualisasi yang menunjukan popularitas relatif dari menu makanan tersebut.
Sebutkan menu makanan yang harus disingkirkan.


```python
import matplotlib.pyplot as plt

# Data menu makanan dan jumlah penjualan
masakan_padang = {
  'Rendang': 95,
  'Dendeng': 93,
  'Ayam Bakar': 90,
  'Ayam Gulai': 92,
  'Gulai Tunjang': 90,
  'Gulai Pakis': 83,
  'Kikil': 85,
  'Ikan Tongkol Balado': 89,
  'Telor Balado': 87,
  'Perkedel': 84
}

# Pisahkan data ke dalam list
menu_makanan = list(masakan_padang.keys())
jumlah_penjualan = list(masakan_padang.values())

# Buat plot
plt.figure(figsize=(10, 6)) #  untuk membuat sebuah figure (gambar) baru dengan ukuran 10 inci lebar dan 6 inci tinggi.
plt.barh(menu_makanan, jumlah_penjualan, color='skyblue')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Menu Makanan')
plt.title('Popularitas Menu Makanan di Rumah Makan Padang')
plt.gca().invert_yaxis()  # Membalikkan sumbu y untuk menampilkan yang paling populer di bagian atas
plt.show()
```

Selanjutnya kita akan mengerjakan tugas 2 yaitu kita akan mengidentifikasi 3 menu makanan yang paling tidak populer dan memberikan alasannya. Mari kita lakukan ini dengan menggunakan Pandas untuk memanipulasi data:

```python
import pandas as pd

# Buat DataFrame dari dictionary
df = pd.DataFrame(list(masakan_padang.items()), columns=['Menu Makanan', 'Jumlah Penjualan'])

# Urutkan DataFrame berdasarkan jumlah penjualan
df_sorted = df.sort_values(by='Jumlah Penjualan')

# Ambil 3 menu makanan yang paling tidak populer
menu_tidak_populer = df_sorted.head(3)

print("Menu makanan yang harus disingkirkan dan alasannya:")
for index, row in menu_tidak_populer.iterrows():
    print(f"{row['Menu Makanan']}: Kurang diminati ({row['Jumlah Penjualan']} penjualan)")
```

Sedangkan 3 menu makanan populer teratas adalah

```python
# Ambil 3 menu makanan terpopuler
menu_terpopuler = df_sorted.tail(3)

print("Menu makanan yang paling populer:")
for index, row in menu_terpopuler.iterrows():
    print(f"{row['Menu Makanan']}: Diminati ({row['Jumlah Penjualan']} penjualan)")
```

Kedai Salgits memiliki berbagai pilihan menu camilan. Untuk menyesuaikan selera konsumen, manajemen ingin mengurangi 3 menu yang kurang laris. Berikut data penjualan menu dalam seminggu terakhir:

Menu_salgits = {

'Salad Buah':65,

'Salad Sayur':60,

'Cireng Ayam':70,

'Cireng Kornet':67,

'Cireng Keju':65,

'Cireng Sosis':58,

'Cireng Bakso':56,

'Jasuke':55,

'Corndog Manis':61,

'Corndog Asin':59 }

Buatlah visualisasi untuk mengetahui popularitas relatif setiap menu. Sebutkan 3 menu yang perlu dihapus berdasarkan data tersebut.

Tugas

Chart apa yang anda pilih untuk problem diatas dan mengapa anda memilih chart tersebut?
Menu Makanan apa saja yang perlu anda sarankan untuk dikeluarkan dari menu?

Untuk masalah ini, saya akan memilih diagram batang horizontal, karena ini adalah cara yang efektif untuk membandingkan kuantitas atau jumlah penjualan dari beberapa item yang berbeda, dalam hal ini menu camilan. Dengan menggunakan diagram batang horizontal, kita dapat dengan jelas melihat perbandingan popularitas relatif dari setiap menu camilan.

Sekarang, mari kita buat visualisasi dan identifikasi menu yang perlu dihapus:

```python
import matplotlib.pyplot as plt

# Data menu camilan dan jumlah penjualan
menu_salgits = {
    'Salad Buah': 65,
    'Salad Sayur': 60,
    'Cireng Ayam': 70,
    'Cireng Kornet': 67,
    'Cireng Keju': 65,
    'Cireng Sosis': 58,
    'Cireng Bakso': 56,
    'Jasuke': 55,
    'Corndog Manis': 61,
    'Corndog Asin': 59
}

# Pisahkan data ke dalam list
menu_camila = list(menu_salgits.keys())
jumlah_penjualan = list(menu_salgits.values())

# Buat plot
plt.figure(figsize=(10, 6))
plt.barh(menu_camila, jumlah_penjualan, color='lightgreen')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Menu Camilan')
plt.title('Popularitas Menu Camilan di Kedai Salgits')
plt.gca().invert_yaxis()  # Membalikkan sumbu y untuk menampilkan yang paling populer di bagian atas
plt.show()

# Identifikasi 3 menu camilan yang paling tidak populer
menu_tidak_populer = sorted(menu_salgits, key=menu_salgits.get)[:3]

print("Menu camilan yang perlu dihapus:")
for menu in menu_tidak_populer:
    print(menu)
```
